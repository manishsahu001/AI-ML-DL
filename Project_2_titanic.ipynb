```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
# Checking columns
df.columns
```

```python
df.info()
```

```python
df.columns
# we have to drop the unused columns from the dataset.
df.drop(['who', 'adult_male', 'deck','embark_town', 'alive',  'class'], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
# checing null values
df.isnull().sum()
```

```python
# Fill the age using mean
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# We have two null values in the embarked column, we can feel this using mean, median and mode or we can remove the rows
# We are dropping the null values.
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.info()
```

```python

df.head()
```

```python
# Label encoding for non int value using sklearn
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
# label encoding for the sex column
df['sex'] = encoder.fit_transform(df['sex'])
```

```python
# label encoding for the embarked column
df['embarked'] = encoder.fit_transform(df['embarked'])
```

```python
df.head()
```

```python
# Converting alone column from boolean to integer
df = df.astype(int)
```

```python
df.head()
```

```python
# Here in this dataset we want to check the survival rate, which we can say that the output and rest of the things are input.
X = df.drop('survived', axis=1)
y = df['survived']
```

```python
# Now we need to do train the model and test the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# Logistic Regression

```python
# We need to use Logistic Regression for the model so we need to import that from the sklearn
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Training the model with input data
model.fit(X_train, y_train)
```

```python
# now we need to do prediction same process
y_pred = model.predict(X_test)
y_pred
```

```python
# here we have 0 and 1 values so we cannot use the R^2 test for this we need to use confusion matrix to test the model accuracy
# To evaluate the model we use confusion matrix
# For the confusion matric model evaluation we will use the sklearn library
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Checking accuracy score
accuracy_score(y_test, y_pred)
```

```python
# Checking confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

# KNN Model

#### feature Scaling

```python
df
```

```python
# Import StandardScaler for the
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# We need to to standard scalling to the age, pclass and fare because we are having the big values in it, we can directly pass the entire dataset, if there is 0 and 1 then it will remain the same and rest of the big values will be scaled.
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# After applying standard scalling we will use the KNN model
from sklearn.neighbors import KNeighborsClassifier
# creating model, we can pass the k value inside this algorithm let's take it as 5
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
# Training the model
knn_model.fit(X_train_scaled, y_train)
```

```python
y_pred_knn_model = knn_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred_knn_model)
```

```python
confusion_matrix(y_test, y_pred_knn_model)
```

```python
print(classification_report(y_test, y_pred_knn_model))
```
